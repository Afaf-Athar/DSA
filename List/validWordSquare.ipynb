### validWordSquare

```python
def validWordSquare(words):

    result = []
    for i in range(len(words[0])):  
        new_word = ''
        for word in words:
            if i < len(word):  
                new_word += word[i]
        result.append(new_word)
    # print(result)
    
    res = []
    for i in words:
        exist =  i in result
        res.append(exist)
        # print(res)

#     for i in res:
#         if i == False:
#             return False
    
#     return True
    for i in range(len(words)):
        # print(f"Comparing word: {words[i]} with result: {result[i]}")
        if words[i] != result[i]:
            return False
    return True
```
