##### Given a list if integers, find the max number from the list

```python

def large_wo_inbuiltfunction(list_large):
    max_num = list_large[0]
    for i in  list_large:
        if max_num < i:
            max_num = i

    return max_num
```

```python
def large_w_inbuiltfunction(list_large):
    list_large.sort(reverse=False)
    max_num = list_large[-1]

    return max_num
```

```python
# list_large = [10,20,4]
# list_large = [10,20,20,4]
list_large = [-10,-20,-20,-4]
large_wo_inbuiltfunction = large_wo_inbuiltfunction(list_large)
large_w_inbuiltfunction = large_w_inbuiltfunction(list_large)
large_wo_inbuiltfunction,large_w_inbuiltfunction
```

##### Given a list of numbers, move all the zeros to one

```python

def zeros_to_end(list_large):
    num = 0
    for i in range(0,len(list_large)):
        if list_large[i] == num:
            list_large.pop(i)
            list_large.append(0)
            
    
    return list_large
```

```python
list_large = [4,0,3,1,0,5,6]
zeros_to_end = zeros_to_end(list_large)
zeros_to_end
```

##### Max Consecutive Ones

```python
def findMaxConsecutiveOnes(nums):
        count = 0
        maxCount = 0
        for i in nums: 
                if i == 1:
                        count = count +1 #
                        maxCount = max(count,maxCount) #1,0
                else:
                        count = 0

        return maxCount
```

```python
# nums =  [1,1,0,1,1,1]
nums = [1,0,1,1,0,1]
findMaxConsecutiveOnes = findMaxConsecutiveOnes(nums)
findMaxConsecutiveOnes
```

##### Find Numbers with Even Number of Digits

```python

def findNumbers(nums):
    even_list = []
    
    for each in nums:
        count = 0
        while(each!=0):
            each = each // 10
            count = count +1

        if count % 2 == 0:
            even_list.append(each)
        # print(l)
    return len(even_list)
            
```

```python
# nums = [555,901,482,1771]
nums = [12,345,2,6,7896]
findNumbers = findNumbers(nums)
findNumbers
```

##### Find the occurence of digit in a given list of numbers

```python

def findNumbersOccurence(nums):
    count = 0
    for each in nums:
        
        while(each!=0):
            # print(each)
            mod = each % 10  
            # print(mod)              
            if mod ==7:
                count = count +1
            each = each//10
            # print(each)
            # print(count)
    return count
            
```

#####  Squares of a Sorted Array -- Heap

```python
def sortedSquares(nums):
    squared_list = []
    for each in nums:
        square = each * each
        squared_list.append(square)
        
    squared_list.sort()
    return squared_list
```

```python
# nums = [-7,-3,2,3,11]
nums = [-4,-1,0,3,10]
sortedSquares = sortedSquares(nums)
sortedSquares
```

```python

def duplicateZeros(arr):
    last = len(arr) - 1
    i = 0
    while(i<=last):     
        if arr[i] == 0:
            k=last
            while(k>i):
                arr[k] = arr[k-1]
                k = k - 1
            i = i+1
        i=i+1
    return arr
            
```

```python
arr = [1,0,2,3,0,4,5,0]
# arr = [1,0,_,2,3,0,4,5]
# arr = [1,0,0,2,3,0,4,5]
# arr = [1,0,2,0,22,2,2]
duplicateZeros = duplicateZeros(arr)
duplicateZeros
```
