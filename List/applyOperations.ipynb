```python

def applyOperations(nums):
    for i in range(len(nums)-1):
        if(nums[i]==nums[i+1]):
            nums[i] = nums[i]*2
            nums[i+1]=0

    non_zero = []
    c = 0
    for i in nums:
        if (i!=0) :
            non_zero.append(i)
        else:
            c +=1 

    return non_zero + [0]*c
```
