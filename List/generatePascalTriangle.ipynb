#### generatePascalTriangle

```python
def generatePascalTriangle(numRows):
    i = 0
    j = 0
    inside_list = []
    pascalList = []
    while((i<numRows) and (j< numRows)):
        for i in range(numRows):
            inside_list.append(1)
            i = i+1
            for j in range(numRows):
                inside_list.append(1)
                j = j + 1
    pascalList.append(inside_list)
    return pascalList
```

```python
numRows = 5
generatePascalTriangle = generatePascalTriangle(numRows)
generatePascalTriangle 
```
