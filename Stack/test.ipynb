```python
#Write a Python code that takes a string and reverses the words in it. For example, 'Hello World' becomes 'World Hello'.Input: 'The quick brown fox'",
```

```python
string = 'The quick brown fox'
len(string)
len(string)
```

```python
def reverse_words_alt(text):
    words = text.split()
    return ' '.join(reversed(words))

reverse_words_alt(string)
```
