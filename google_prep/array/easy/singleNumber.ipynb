136. Single Number

https://leetcode.com/problems/single-number/?envType=problem-list-v2&envId=xleo1moc

```python

def singleNumber(nums):

    seen = {}
    count = 1
    for i in nums:
        if i in seen:
            seen[i] = count+1
        else:
            seen[i] = count
    print(seen)

    for key,value in seen.items():
        if (value==1):
            return key
    
 
```

```python
# nums = [2,2,1]
nums = [4,1,2,1,2]
singleNumber = singleNumber(nums)
singleNumber
```
