217. Contains Duplicate

https://leetcode.com/problems/contains-duplicate/description/?envType=problem-list-v2&envId=xleo1moc

```python

def containsDuplicate(nums):

    dic = {}
    for num in nums:
        if num not in dic:
            dic[num] = 1
        else:
            dic[num] += +1
    
    # print(dic)
    for key,value in dic.items():
        if(value>1):
            return True
    return False
```

```python
# nums = [4,3,2,7,8,2,3,1] # [5,6]
# nums = [1,1]
nums = [1,2,3,4]
containsDuplicate = containsDuplicate(nums)
containsDuplicate
```
