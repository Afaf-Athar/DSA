128. Longest Consecutive Sequence

https://leetcode.com/problems/longest-consecutive-sequence/?envType=problem-list-v2&envId=xleo1moc

```python

def longestConsecutive(nums):

    i = 0
    j = 1
    count = 1
    max_count = 0
    
    if len(nums)==0:
        return 0
    if len(nums)==1:
        return 1

    sort_arr = sorted((nums))
    print(sort_arr)

   
    while(j<len(sort_arr)):
        if((sort_arr[j]==sort_arr[i]+1) and (sort_arr[j]!=sort_arr[i])):
            
            i +=1
            j +=1
            count +=1 
            max_count =  max(max_count,count)
            
        else:
            count = 1
            i = j
            j = j+1
            max_count =  max(max_count,count)

    return max_count
```
