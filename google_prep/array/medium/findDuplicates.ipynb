442. Find All Duplicates in an Array

https://leetcode.com/problems/find-all-duplicates-in-an-array/description/?envType=problem-list-v2&envId=xleo1moc

```python

def findDuplicates(nums):
    if len(nums)==1:
        return []
    
    seen = set()
    dup = []
    for i in nums:
        if i in seen:
            dup.append(i)
        else:
            seen.add(i)
    
    return dup

    
    
```

```python
# nums = [4,3,2,7,8,2,3,1]
nums = [1,1,2]
# nums = [1]

findDuplicates = findDuplicates(nums)
findDuplicates
```
