##### deleteDuplicates

```python
class ListNode:
    def __init__(self,val) :
        self.val = val
        self.next = None


def deleteDuplicates(head):
    if (head == None or head.next == None):
            return head
            
    tempHead = head
    headnext = tempHead.next

    while(headnext!=None):


        if(tempHead.val != headnext.val):
            tempHead.next = headnext
            tempHead = tempHead.next
            headnext = headnext.next
        else:
        #    tempHead = headnext
            headnext = headnext.next

    tempHead.next = None

    return head
    
        


def createlinkedlist(index,nums):
    if (index==len(nums)):
        return None
    
    node = ListNode(nums[index])
    node.next = createlinkedlist(index+1,nums)

    return node


```

```python
# nums = [1,1,1,2,3]
nums = [1,1,2,3,3]

firstNode = createlinkedlist(0,nums)
deleteDuplicated = deleteDuplicates(firstNode)
deleteDuplicated

while(deleteDuplicated!=None):
    print(deleteDuplicated.val)
    deleteDuplicated = deleteDuplicated.next
```
