```python
class singleNode:

    def __init__(self,val):
        self.val = val
        self.ptr = None
        
def show_list(head):
    
    while(head != None):
        print(head.val)
        head = head.ptr
        
def create_list(nums):
    head = link(0,nums)
    return head


def link(index,nums):
    if(index == len(nums)):
        return None
    
    node = singleNode(nums[index])
    node.ptr = link(index+1,nums)

    return node

```

```python
nums = [95,96,97,43,45,32,98]
create_list = create_list(nums)
show_list(create_list)
# created_list = show_list(nums)
# created_list
# print(created_list.val)
# sec = created_list.ptr
# print(sec.val)
```
