```python
def sortedArrayToBST(nums):
    nums_len = len(nums)
    mid = (nums_len)//2
    i = mid-1
    j = len(nums)-1
    binary_list = []
    

    if len(nums)==0:
        return None
    
    while(i>=0 and j < nums_len):
        node = nums[mid]
        binary_list.append(node)
        
        if (nums[i]< node & (nums[i]!=0)):
            binary_list.append(nums[i])
            i = i-1
            while((len(binary_list)-1)<j):
                binary_list.append(nums[j])
                
                j = j-1
                
    
    return binary_list

    
nums = [-10,-3,0,5,9]
sortedArrayToBST = sortedArrayToBST(nums)
sortedArrayToBST
```
